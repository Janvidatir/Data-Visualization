## Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df.head()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```
